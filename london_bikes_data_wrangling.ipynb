```python
import pandas as pd
```

```python
london_bikes = pd.read_csv("london_merged.csv")
```

```python
london_bikes
```

```python
new_col_names = {
    "timestamp": "time",
    "cnt": "count",
    "t1": "temp_C",
    "t2": "temp_feels_like_C",
    "hum": "humidity",
    "wind_speed": "wind_speed_kph",
    "weather_code": "weather",
    "is_holiday": "is_holiday",
    "is_weekend": "is_weekend",
    "season": "season"
}

london_bikes.rename(columns=new_col_names, inplace=True)
```

```python
london_bikes["humidity"] = london_bikes["humidity"].apply(lambda x: x / 100)
london_bikes["season"] = london_bikes["season"].astype(int)
london_bikes["weather"] = london_bikes["weather"].astype(int)
```

```python
season_codes = {
    0: "Spring",
    1: "Summer",
    2: "Fall",
    3: "Winter"
}

weather_codes = {
    1: "Clear",
    2: "Scattered clouds",
    3: "Broken clouds",
    4: "Cloudy",
    7: "Rain",
    10: "Rain with thunderstorm",
    26: "Snowfall",
    94: "Freezing Fog"
}

london_bikes["season"] = london_bikes["season"].map(season_codes)
london_bikes["weather"] = london_bikes["weather"].map(weather_codes)
```

```python
london_bikes.to_csv("london_bikes.csv", index=False)
```
